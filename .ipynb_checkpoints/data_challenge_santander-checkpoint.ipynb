# Web Scraping 
Obtendo os dados do site Reclame Aqui

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains

from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup as bs
import csv

import time
import timeit

import pandas as pd
from datetime import datetime
```

```python
def obtem_detalhes(link, driver):
    
    try:
        driver.get(link)

        time.sleep(4)

        html = bs(driver.page_source, "html.parser")

        comentario_head = html.find('div', {'class': 'complain-head'})
        if comentario_head is not None:
            titulo_elemento = comentario_head.find('h1')
            titulo = titulo_elemento.text

            ul = comentario_head.find('ul', {'class': 'local-date'})
            li = ul.find_all('li')
            local = li[0].text
            ID = li[1].text
            data = li[2].text
        else:
            titulo = '',
            local = '',
            ID = '0',
            data = ''


        comentario_body = html.find('div', {'class': 'complain-body'})
        if comentario_body is not None:
            comentario_completo = comentario_body.find('p').text
        else:
            comentario_completo = ''
        
        return titulo, comentario_completo, local, ID, data
    
    except:
        return '', '', '', '0', ''
    
```

```python
def obtem_reclamacoes():
    
    data_inicial = datetime.strptime('01/04/20', "%d/%m/%y")
    
    driver = webdriver.Chrome('./chromedriver')
    driver.get('https://www.reclameaqui.com.br/busca/?q=cal%C3%A7ados')
    
    time.sleep(3)
    
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    
    driver_detalhe = webdriver.Chrome('./chromedriver', options=chrome_options)
    
    time.sleep(3)
    
    while True:
    
        html = bs(driver.page_source, "html.parser")

        items = html.find_all("div", {"class": "complain-item"})

        reclamacoes = []

        for div in items:
            titulo_element = div.find('h2')
            titulo = titulo_element.text if titulo_element is not None else ''
            empresa_element = div.find('span', {'class': 'office-building-text'})
            nome_empresa = empresa_element.text if empresa_element is not None else ''
            comentario_element = div.find('p', {'class': 'complaint-content'})
            comentario_curto = comentario_element.text if comentario_element is not None else ''
            topo = div.find('span', {'class': 'complain-status-top'})
            if topo is not None:
                link_descricao_completa = topo.find_all('a')
                if link_descricao_completa is not None and len(link_descricao_completa) > 0:
                    links = link_descricao_completa[0]
                    link = 'https://www.reclameaqui.com.br' + links['href']
            else:
                link = ''
            data_element = div.find('em', {'class': 'mdi-calendar-today'})
            data_desc = data_element.next_element.strip() if data_element is not None else ''
            cidade_element = div.find('span', {'class': 'mdi-map-marker'})
            cidade = cidade_element.next_element.next_element.text if cidade_element is not None else ''

            titulo_completo, comentario_completo, local, ID, data = obtem_detalhes(link, driver_detalhe)
 
            _id = ID.replace('ID: ', '')
            
            with open('dataset_junho.csv', mode='at') as dataset_file:
                writer = csv.writer(dataset_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                writer.writerow([_id, titulo, titulo_completo, nome_empresa, comentario_curto, comentario_completo, link, data, cidade, local])
        
        data_reclamacao_str = data.split('às')[0].strip()
        data_reclamacao = datetime.strptime(data_reclamacao_str, "%d/%m/%y")
        
        botao_proximo = html.find('li', {'class': 'pagination-next ng-scope disabled'})
        
        print(data_reclamacao, data_inicial, botao_proximo)
        if data_reclamacao <= data_inicial or botao_proximo is not None:
            break     
        
        driver.find_element_by_css_selector("a.ng-binding[ng-click^='selectPage(page + 1, $event)']").click()
       
        time.sleep(3)        
        
    driver_detalhe.close()  
    driver.close()
```

# Processamento de linguagem natural

```python
# conda install -c conda-forge spacy
# !python -m spacy download pt_core_news_sm
```

```python
import spacy
nlp = spacy.load("pt_core_news_sm")
```

```python
from spacy.lang.pt.stop_words import STOP_WORDS
```

```python
# STOP_WORDS definidas pelo grupo de acordo com o problema em questão:
novas_stop_words = ['produto', 'produtos', 'compra', 'calçado', 'entregue', 'calçados', 'empresa', 'problema']
```

## Pré-processamento dos títulos e comentários

```python
titulos_sem_stop_words = []

for titulo in data_frame['titulo']:
    doc = nlp(titulo)
    texto = []
    for token in doc:
        if not token.is_punct | token.is_stop | token.is_space:
            if token.text.lower() not in novas_stop_words:
                texto.append(token.text.lower())
            
    titulo = ' '.join([palavra for palavra in texto])
    titulos_sem_stop_words.append(titulo)
    
data_frame['titulo_preprocessado'] = titulos_sem_stop_words
```

```python
comentarios_processados = []


for comentario in data_frame['comentario']:
    doc = nlp(comentario)
    comentario_palavras = []
    for token in doc:
        if not token.is_punct | token.is_stop | token.is_space and token.text not in novas_stop_words:
            comentario_palavras.append(token.text.lower())
            
    texto = ' '.join([palavra for palavra in comentario_palavras])
    comentarios_processados.append(texto)
    
data_frame['comentario_preprocessado'] = comentarios_processados
```

```python
documents = list()

for sentenca in data_frame['comentario_preprocessado']:
    documents.append(list(gensim.utils.simple_preprocess (sentenca)))
```

### Análise considerando tipos de problemas

```python
import pandas as pd
import numpy as np

# Abre data frame com todas as incidencias de junho e julho
df_completo = pd.read_csv(r'dataset_jul_jun.csv')

# Nome dos arquivos a serem abertos para geração do data frame final
lista_arqs = ['dataset_problemas_3',
         'dataset_problemas_0',
         'dataset_problemas_2']

# Inicia um data frame vazio para dar append nos data frames que serão abertos
df = pd.DataFrame()

    
# Remove ID com erro no scrapping
df_completo = df_completo[df_completo.ID != ' 16/07/20 às 19h53']
# Transformando os ID's para inteiros
df_completo.ID = df_completo.ID.astype('int')

# Looping para abrir os arquivos extraídos e adiciona-los a um data frame
for arquivo in lista_arqs:
    df_appender = pd.read_csv(arquivo + '.csv')
    df = df.append(df_appender)

# Transformando os ID's para inteiros
df.ID = df.ID.astype(int)

# Mescla data frames para obter ID's que possuem registro de problemas 
df_final = pd.merge(df_completo, df, how='left',
                    on = ['ID', 'titulo', 'titulo_completo', 'empresa', 'comentario_curto',
       'comentario', 'link', 'data', 'cidade', 'local'])
# Remove registros duplicados
df_final.drop_duplicates(subset=['ID'], inplace = True)

# Converte string das colunas para minusúculo
df_final['titulo_completo'] = df_final['titulo_completo'].str.lower()
df_final['comentario'] = df_final['comentario'].str.lower()

# Geral um mapping para classificar os problemas
problema_map = {0: 'Má qualidade',
                2: 'Problema com entrega',
                3: 'Devolução ou Troca'}

# Mapeia os problemas
df_final['problema'] = df_final['problema'].map(problema_map)


# Inicia listas de problemas a serem classificados
problemas = ['Devolução ou Troca',
            'Má qualidade',
             'Problemas com entrega',
             'Atendimento/Contato',
             'Propaganda enganosa',
             'Juros/Taxas']

# Inicia-se condições para verificar se o título possui uma determinada
# substring e, por consequência, classifica-lo em algum dos problemas 
defeito_t = df_final['titulo_completo'].str.contains('defeito|qualidade')
atraso_t = df_final['titulo_completo'].str.contains('atras|entreg|recebi')
devolução_troca_t = df_final['titulo_completo'].str.contains('troca|devolução')
contato_t = df_final['titulo_completo'].str.contains('contato|atendimento')
propaganda_t = df_final['titulo_completo'].str.contains('enganos|propaganda')
taxas_juros_t = df_final['titulo_completo'].str.contains('juro|taxa|indevid')

# Adiciona coluna relatando problema a partir do título
# caso um valor não seja achado, a coluna guardará o valor da coluna 'problema'
df_final['problema_analise_t'] = np.select([devolução_troca_t, defeito_t, atraso_t, contato_t, propaganda_t, taxas_juros_t],
                                           problemas, df_final['problema'])


# Inicia-se condições para verificar se o comentário possui uma determinada
# substring e, por consequência, classifica-lo em algum dos problemas 
defeito_c = df_final['comentario'].str.contains('defeito|qualidade|sola')
atraso_c = df_final['comentario'].str.contains('atras|entreg|recebi')
devolução_troca_c = df_final['comentario'].str.contains('troca|devolução')
contato_c = df_final['comentario'].str.contains('contato|atendimento')
propaganda_c = df_final['comentario'].str.contains('enganosa|propaganda')
taxas_juros_c = df_final['comentario'].str.contains('juro|taxa|indevid')

# Adiciona coluna relatando problema a partir do comentário
# caso um valor não seja achado,
# a coluna guardará o valor da coluna 'problema_analise_t'
df_final['problema_analise_c'] = np.select([devolução_troca_c, defeito_c, atraso_c, contato_t, propaganda_c, taxas_juros_c],
                                           problemas, df_final['problema_analise_t'])


# Condições para determinar a análise final do problema
# dessa forma, o grau de determinação da análise final baseia-se na ideia de que
# o problema do scrapping é mais acertivo, em sequência a análise feita no
# título e depois a do comentário
problemas_condicoes = [~df_final['problema'].isnull(),
                       ~df_final['problema_analise_t'].isnull(),
                       ~df_final['problema_analise_c'].isnull()]

# Ordem de relevância das análises
problemas_input = [df_final['problema'],
                   df_final['problema_analise_t'],
                   df_final['problema_analise_c']]

# Caso nenhuma das análises seja encontrada, o problema é classificado como
# 'Outros problemas'
df_final['problema_final'] = np.select(problemas_condicoes,
                                       problemas_input, 'Outros problemas')
# Adiciona coluna de estado ao data frame
df_final['Estado'] = df_final.local.str[-2:]
# Filtra possíveis problemas do scrapping
df_final = df_final[~df_final['Estado'].str.contains('-')]


# Exporta data frame
df_final.to_excel('dataset_problema_categorizado.xlsx',
                  index = False)
```

```python
# Leitura do dataset criado anteriormente, tendo sido definido suas categorias de acordo com o site ReclameAqui.
df_final = pd.read_excel('dataset_problema_categorizado.xlsx')
```

```python
'''                GROUPED BAR CHART ESTADO VS PROBLEMAS                   '''

# Gera colormap para os problemas
cmap = ListedColormap(sns.color_palette("coolwarm", 7))
# Cruza estado com tipo de problema para analisar frequências
estado_problema = pd.crosstab(df_final['Estado'],
                df_final['problema_final'])

# Normaliza dados para que estejam as amostragens de linha representem
# 100% dos dados em procentagem
estado_problema = estado_problema.div(estado_problema.sum(1) / 100, axis=0)

# Plota stacked horizontla bar chart de Estados por tipo de problema
fig, ax = plt.subplots(figsize=(len(estado_problema)/1.25, 12.5))
estado_problema.plot(kind='barh', stacked=True, ax = ax,  cmap = cmap)

# Anota porcentagens nas barras
for rec in ax.patches:
    height = rec.get_height()
    width = rec.get_width()
    if width > 0:
        ax.text(rec.get_x() + rec.get_width() / 2, 
                  rec.get_y() + height / 2,
                  "{:.0f}%".format(width),
                  ha='center', 
                  va='center',
                  fontsize=10, color='black')
        
# Formata legenda
plt.legend(bbox_to_anchor=(.675, -0.05), ncol=3)

# Desliga spines ('grids')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove x ticks
ax.set_xticks([])
# Rotaciona Y ticks labels
plt.yticks(rotation = 25)
# plota o gráfico
plt.show()
```

```python

'''                    Gráfico de barras problemas em SP                   '''

# Cria um novo dataframe filtrando pela cidade de são paulo
df_sp = df_final[df_final.cidade == 'São Paulo']
# Cria series mostrando frequências de cada problema
df_sp_count = df_sp['problema_final'].value_counts().rename_axis('Problemas')
# Reordena os dados
df_sp_count.sort_values(inplace = True)

# Plota gráfico de barras mostrando a representatividade dos problemas
# enfrentados pelos clientes de São Paulo
fig, ax = plt.subplots(figsize=(8, 4))
df_sp_count.plot(kind='barh', ax = ax, color='#8f000a',
                 title = 'Problemas nas compras de calçados em São Paulo (SP)')
# Armazena total de casos
soma_count = df_sp_count.sum()
# Anota quantidade total do problema e sua representação no todo
for y, x in enumerate(df_sp_count.values):
    plt.annotate(' {} incidentes ({}%)'.format(x, int(x/soma_count * 100)), xy=(x, y), va='center')
    
# Desliga spines ('grids')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove x ticks
ax.set_xticks([])
# Remove y labels
plt.ylabel('')
# Remove ticks marker
plt.tick_params(
    axis='y',          
    which='both',      
    left=False,      
    labelleft=True)
# plota o gráfico
plt.show()
```

```python

'''                Subplots das top empresas com mais reclamações           '''

# Função para anotar as representações na pie, caso sejam maiores que 0
def auto_pct_fill(pct):
    if pct > 0:
        return '{:d}%'.format(int(pct))
    else:
        return ''

# Armazena nome das empresas com mais reclamações
top_5_sp = df_sp.empresa.value_counts().head(5).index.tolist()
# Cria um novo data frame filtrando pelo nome dessas empresas
df_sp_top5_problemas = df_sp[df_sp['empresa'].isin(top_5_sp)]
# Cruz nome da empresa com tipo de problema para obter frequências
emp_problema = pd.crosstab(df_sp_top5_problemas['empresa'],
                                  df_sp_top5_problemas['problema_final'])

# Determina paleta de cores do plot
color_muted = sns.color_palette("RdBu", 7)
# Transforma paleta em hexcodes
color_muted.as_hex()

# Plota subplots de pie charts mostrando a representatividade dos problema
# por empresa
fig, ax = plt.subplots(3,2)
fig.set_size_inches(7.5,10)
for empresa, axf in zip(top_5_sp, ax.flatten()):
    # Filtra por empresa
    df_empresa = emp_problema[emp_problema.index ==  empresa]
    
    # armazena frequências
    data_problema = df_empresa.values.tolist()
    # armazena label dos problemas
    labels_problema = df_empresa.columns.tolist()
    
    # Plota pie chart no axis do subplot
    axf.pie(data_problema, labels=labels_problema,
        autopct=lambda pct: auto_pct_fill(pct), colors = color_muted,
        labeldistance=None,
        textprops=dict(color="black", size=11))
    # Determina titulo do subplot
    axf.set_title(empresa)
    # Adiciona legenda
    handles, labels = axf.get_legend_handles_labels()
    fig.legend(handles, labels, loc='lower right',  bbox_to_anchor=(0.75, .075))

# Remove subplot inutilizado
ax[-1, -1].axis('off')
# Determina suptitle dos subplots
titulo = 'Distribuição de problemas nas TOP 5 empresas\n com mais registros de reclamações'
fig.suptitle(titulo)
# plota o gráfico
plt.show()
```
